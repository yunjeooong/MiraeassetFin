```python
import pandas as pd
```

```python
df1 = pd.read_excel('/content/naver_news_content_with_answers_1_2_투자레터_경제만.xlsx')
df1
```

```python
df2= pd.read_excel('/content/상장종목.xlsx')
df2
```

```python
# 관련주 컬럼 추가
def extract_related_stock(answer, stock_list):
    related_stocks = [stock for stock in stock_list if stock in answer]
    return ', '.join(related_stocks)

df1['관련주'] = df1['답변'].apply(lambda x: extract_related_stock(x, df2['corp_name'].tolist()))

df1
```

```python
df1 = df1[df1['관련주'] != '']
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
# 중복된 내용 삭제
def remove_duplicates(related_stocks):
    seen = set()
    unique_stocks = []
    for stock in related_stocks.split(', '):
        if stock not in seen:
            seen.add(stock)
            unique_stocks.append(stock)
    return ', '.join(unique_stocks)

df1['관련주'] = df1['관련주'].apply(remove_duplicates)
```

```python
df1
```

```python
# '관련주'에서 종목별 빈도 계산
all_related_stocks = df1['관련주'].str.split(', ').explode()
stock_counts = Counter(all_related_stocks)

# 많은 순으로 정렬된 결과
sorted_stock_counts = stock_counts.most_common()
```

```python
# 상위 5개 추출
top_5_stocks = sorted_stock_counts[:5]
top_5_stocks
```

```python
# 종목명만 추출하여 데이터프레임 생성
top_5_stock_names = [stock for stock, count in top_5_stocks]
top_5_df = pd.DataFrame(top_5_stock_names, columns=['top5종목'])
```

```python
top_5_df
```

```python
top_5_df.to_excel('경제_top5종목.xlsx', index=False)
```

```python
# '관련주' 컬럼에 top_5_stock_names가 포함된 행 필터링
def contains_any_related_stock(related_stocks, stock_names):
    stocks = related_stocks.split(', ')
    return any(stock in stock_names for stock in stocks)

filtered_df = df1[df1['관련주'].apply(lambda x: contains_any_related_stock(x, top_5_stock_names))]
filtered_df.reset_index(drop=True, inplace=True)
filtered_df
```

```python
filtered_df.to_excel('경제_top5종목_뉴스.xlsx', index=False)
```
