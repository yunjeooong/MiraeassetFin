```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이터 입력
dates = pd.date_range(start="6/1/2024", end="6/30/2024")
quantities = [3377, 8513, 20196, 21290, 16715, 10881, 13097, 3650, 9263, 20975,
              20960, 21054, 21376, 14717, 3698, 9577, 21296, 21622, 20906, 21610,
              15110, 3780, 9666, 21165, 21593, 20976, 22349, 15221, 4073, 10312]

# DataFrame 생성
data = pd.DataFrame({'Date': dates, 'Quantity': quantities})

# 시각화
plt.figure(figsize=(10, 6))
plt.bar(data['Date'], data['Quantity'], color='navy')
plt.title('Daily Quantities in June 2024')
plt.xlabel('Date')
plt.ylabel('Quantity')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```

```python
import statistics

# 평균 계산
average_quantity = statistics.mean(quantities)

print(f'The average quantity is: {average_quantity}')
```
